```python
power = [12.9,23.7,31.2,33.1,49.4,62.7,69.3]

data = [f'Cut Videos/{p}x200,t=0_trapped.csv' for p in power]

MSD_x_list,MSD_y_list,equip_x_list,equip_y_list = [],[],[],[]

for file_name in data:
    print(f"processing: {file_name}")
    MSD_x,MSD_y,equip_x,equip_y = process_data(file_name, plot_figures=True)
    MSD_x_list.append(MSD_x)
    MSD_y_list.append(MSD_y)
    equip_x_list.append(equip_x)
    equip_y_list.append(equip_y)
```
